```python
from google.cloud import bigquery
import pandas as pd

project_id = "logistics-customer-staging"
logging.info(f"Initializing BigQuery client for project: {project_id}")

try:
    # Initialize BigQuery client
    client = bigquery.Client(project=project_id)
except Exception as e:
    logging.error(f"Failed to initialize BigQuery client: {e}")
    sys.exit(1)

# Define your project and dataset
PROJECT_ID = project_id
DATASET_ID = "your_dataset_id"
TABLE_ID = "your_table_id"

# Sample DataFrame
data = {
    "id": [1, 2, 3],
    "name": ["Alice", "Bob", "Charlie"],
    "created_at": ["2023-01-01", "2023-01-02", "2023-01-03"]
}
df = pd.DataFrame(data)

# Ensure 'created_at' is in datetime format
df["created_at"] = pd.to_datetime(df["created_at"])

# Define the table reference
table_ref = f"{PROJECT_ID}.{DATASET_ID}.{TABLE_ID}"

# Create the table in BigQuery
job = client.load_table_from_dataframe(
    df,
    table_ref,
    job_config=bigquery.LoadJobConfig(
        write_disposition="WRITE_TRUNCATE"  # Overwrites table if it exists
    )
)
job.result()  # Wait for the job to complete

print(f"Table {TABLE_ID} created successfully in dataset {DATASET_ID}.")
```
